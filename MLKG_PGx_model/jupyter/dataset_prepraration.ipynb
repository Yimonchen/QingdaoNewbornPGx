```python
df_pharmvar = pd.read_csv("static/pharmvar_GRCh38_all.csv", dtype=str).fillna("")
```

```python
df_pharmvar = df_pharmvar[~df_pharmvar["haplotype_name"].str.contains("\.")]
```

```python
df_single = pd.DataFrame(columns=df_pharmvar.columns)
df_multi = pd.DataFrame(columns=df_pharmvar.columns)
for key, content in df_pharmvar.groupby(["haplotype_name"]):
    if len(content) == 1:
        df_single = pd.concat([df_single, content], axis=0)
    else:
        df_multi = pd.concat([df_multi, content], axis=0)
```

```python
df_multi = df_multi.groupby(["haplotype_name", "gene"])["reference_sequence"].count().reset_index(name='count')
df_multi.sort_values(by=["gene", "haplotype_name"])
os.chdir("D:/functional-prediction/")
df_multi.to_csv("output/multi_hap.csv", index=False)
```

```python
df_output = df_single[[
    "reference_sequence", "variant_start", "variant_end", 
    "reference_allele", "variant_allele", "type", "haplotype_name", "gene"
]]
df_output = df_output[df_output["reference_allele"] != ""]
```

```python
df_output = df_output[["gene", "haplotype_name", "chr", "variant_start", "reference_allele", "variant_allele"]].drop_duplicates()
```

```python
dpyd_map_dict = {}
for key, value in function_dict.items():
    if "DPYD" in key:
        mpping =  v.haplotype_mapping(key)
        if "rsID" in mpping.keys():
            rs_list = v.haplotype_mapping(key)["rsID"]
            if len(rs_list) > 0:
                dpyd_map_dict[rs_list[0]] = key
```

```python
function_list = []
for index, row in df_output.iterrows():
    haplotype_name = row["haplotype_name"]
    if "rs" in haplotype_name and haplotype_name in dpyd_map_dict.keys():
        haplotype_name = dpyd_map_dict[haplotype_name]
    
    if haplotype_name in function_dict.keys():
        function_list.append(function_dict[haplotype_name])
        
    else:
        function_list.append("")
        
df_output["function"] = function_list
```

```python
chr_regex = re.compile(r"chromosome [\dX]+,")
pos_regex = re.compile(r"^[\d]+")

table_dict = {
    "gene": [],
    "haplotype_name": [],
    "chr": [],
    "variant_start": [],
    "reference_allele": [],
    "variant_allele": [],
    "function": []
}
    
for key, value in function_dict.items():
    if "UGT1A1" in key:
        gene = "UGT1A1"
    
    elif "RYR1" in key:
        gene = "RYR1"
        
    elif "TPMT" in key:
        gene = "TPMT"
        
    elif "G6PD" in key:
        gene = "G6PD"
    
    else:
        gene = ""
        
    if gene != "":
        h_map = v.haplotype_mapping(key)
        if "NC_list" in h_map.keys() and "NC_column" in h_map.keys():
            if len(h_map["NC_list"]) > 1:
                continue
            
            nc_column = h_map["NC_column"]
            chr_ext = re.findall(chr_regex, nc_column)[0]
            chromo = chr_ext.split(" ")[-1].strip(",")
            
            position = h_map["NC_list"][0]
            position = position.split(".")[-1]
            pos = re.findall(pos_regex, position)[0]
            change = position.replace(pos, "")
            
            if ">" in change:
                ref, alt = change.split(">")
            else:
                ref = ""
                alt = change
                
            if "malignant" in value.lower():
                value = "Increased Function"
            
            table_dict["gene"].append(gene)
            table_dict["haplotype_name"].append(key)
            table_dict["chr"].append(chromo)
            table_dict["variant_start"].append(pos)
            table_dict["reference_allele"].append(ref)
            table_dict["variant_allele"].append(alt)
            table_dict["function"].append(value)
```

```python
df_add = pd.DataFrame(table_dict)
```
